```python
from random import randint, shuffle
from time import perf_counter

def quick_test(quick_sort):
    a = []
    quick_sort(a)
    assert a == []

    a = [1]
    quick_sort(a)
    assert a == [1]

    a = [2,3,4,1]
    quick_sort(a)
    assert a == [1,2,3,4]

    for t in range(100):
        a = list(range(randint(1, 1000)))
        shuffle(a)
        quick_sort(a)
        assert a == sorted(a)

    for t in range(5):
        a = list(range(50000))
        shuffle(a)
        t1 = perf_counter()
        quick_sort(a)
        t2 = perf_counter()
        assert a == sorted(a)
        assert t2-t1 < 1

    print("ok")

def quick_sort1(a):
    subs = [(0, len(a)-1)]
    while subs:
        l, r = subs.pop()
        if l >= r:
            continue

        i = randint(l, r)
        x = a[i]
        a[l], a[i] = x, a[l]
        m = l

        for k in range(l+1,r+1):
            if a[k] <= x:
                m += 1
                a[k], a[m] = a[m], a[k]

        a[l], a[m] = a[m], a[l]

        if m < (l+r)//2:
            subs.extend([(l, m-1), (m+1, r)])
        else:
            subs.extend([(m+1, r), (l, m-1)])


def quick_sort1m(a):
    subs = [(0, len(a)-1)]
    while subs:
        l, r = subs.pop()
        if l >= r:
            continue

        m = (l+r)//2
        i = m if a[l]<=a[m]<=a[r] else l if a[m]<=a[l]<=a[r] else r
        x = a[i]
        a[l], a[i] = x, a[l]
        m = l

        for k in range(l+1,r+1):
            if a[k] <= x:
                m += 1
                a[k], a[m] = a[m], a[k]

        a[l], a[m] = a[m], a[l]

        if m < (l+r)//2:
            subs.extend([(l, m-1), (m+1, r)])
        else:
            subs.extend([(m+1, r), (l, m-1)])


def quick_sort3(a):
    subs = [(0, len(a)-1)]
    while subs:
        l, r = subs.pop()
        if l >= r:
            continue

        m = (l+r)//2
        i = m if a[l]<=a[m]<=a[r] else l if a[m]<=a[l]<=a[r] else r
        p = a[i]

        s = l-1
        e = l
        a[i] = a[e]
        a[e] = p

        for i in range(l+1,r+1):
            v = a[i]
            if v > p:
                continue
            e += 1
            a[i] = a[e]
            a[e] = p
            if v < p:
                s += 1
                a[s] = v

        if s-l < r-e+1:
            subs.extend([(l, s), (e+1, r)])
        else:
            subs.extend([(e+1, r), (l, s)])


def quick_sort3m(a):
    if len(a) <= 1:
        return

    cut, que = (0, len(a)), []
    while cut:
        # next cut: left, right
        l, r = cut

        # start, end, middle
        s, e = l, r-1
        m = (s + e) // 2

        # choose pivot
        i = m if a[s]<=a[m]<=a[e] else s if a[m]<=a[s]<=a[e] else e
        p = a[i]

        # init ranges: smaller, equal
        s, e = l, l+1
        a[i], a[s] = a[s], p

        # partition
        for i in range(e, r):
            v = a[i]
            if v > p:
                continue
            a[i], a[e] = a[e], p
            e += 1
            if v < p:
                a[s] = v
                s += 1

        # arrange ranges
        cut, qcut = (l, s), (e, r)
        if cut[1] - cut[0] < qcut[1] - qcut[0]:
            cut, qcut = qcut, cut

        # recurse by iteration
        if qcut[0] < qcut[1]:
            que.append(qcut)
        if cut[0] >=cut[1]:
            cut = que.pop() if que else None


quick_test(quick_sort3m)
```
